```python
# 날짜 형식 통일
df_fx["date"] = pd.to_datetime(df_fx["date"])
df_ksvkospi["날짜"] = pd.to_datetime(df_ksvkospi["날짜"])
df_261250["날짜"] = pd.to_datetime(df_261250["날짜"])
df_261260["날짜"] = pd.to_datetime(df_261260["날짜"])

# 날짜 컬럼명 통일
df_ksvkospi = df_ksvkospi.rename(columns={"날짜": "date", "종가": "ksvkospi 종가"})
df_261250 = df_261250.rename(columns={"날짜": "date", "종가": "미국달러선물레버리지"})
df_261260 = df_261260.rename(columns={"날짜": "date", "종가": "미국달러선물인버스"})

# 컬럼명 변경
df_ksvkospi = df_ksvkospi.rename(columns={"종가": "ksvkospi 종가"})
df_261250 = df_261250.rename(columns={"종가": "미국달러선물레버리지"})
df_261260 = df_261260.rename(columns={"종가": "미국달러선물인버스"})

# 날짜 기준 병합
df_merged = df_fx.merge(df_ksvkospi[["date", "ksvkospi 종가"]], on="date", how="left")
df_merged = df_merged.merge(df_261250[["date", "미국달러선물레버리지"]], on="date", how="left")
df_merged = df_merged.merge(df_261260[["date", "미국달러선물인버스"]], on="date", how="left")

# 결과 확인
df_merged
```

```python
def eda_basic_info(df, df_name="DataFrame"):
    print(f"\n===== {df_name} =====")
    print("Head (5 rows):")
    display(df.head())
    
    print("\nInfo:")
    print(df.info())
    
    print("\nDescribe (numeric):")
    display(df.describe())
    
    print("\nNull counts:")
    print(df.isnull().sum())
    
    print("="*40)
```

```python
eda_basic_info(df_merged, "df")
```

```python
# '미국달러선물레버리지' 컬럼 삭제
df_merged = df_merged.drop(columns=["미국달러선물레버리지"])

eda_basic_info(df_merged, "df")
```
